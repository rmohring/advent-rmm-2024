```python
data = []
for line in fh.readlines():
    vals = line.strip("\n").split(" ")
    x2 = [int(x) for x in vals]
    data.append(np.array(x2))
```

```python
data
```

```python
blah = []
for row in data:
    diff = np.diff(row)
    c1 = np.prod(np.abs(diff)<=3)
    c2 = np.prod(diff<0)
    c3 = np.prod(diff>0)
    answer = c1*(c2+c3)
    print(answer)
    blah.append(answer)
```
