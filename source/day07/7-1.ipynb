```python
import numpy as np
import pandas as pd
#import rmm_util_accessor
import itertools
import time
import re
import grid
```

```python
day = 7
#infile = f'/home/rmohring/code/rmm/advent-rmm-2024/source/day{day:02}/{day}-1-test.dat'
infile = f'/home/rmohring/code/rmm/advent-rmm-2024/source/day{day:02}/{day}-1.dat'
#infile = f'/home/rmohring/code/rmm/advent-rmm-2024/source/day{day:02}/{day}-2.dat'
data = []
with open(infile, 'r') as fh:
    rawdata = fh.readlines()
```

```python
data = []
for row in rawdata:
    tmp = row.strip('\n')
    ans, eqq = tmp.split(": ")
    ans = int(ans)
    eq = [int(x) for x in eqq.split(" ")]
    data.append( (ans, eq) )
```

```python
data
```

```python
%%time
good = []
for ans, eq in data:
    q = {}
    #print(ans, eq)
    q[0] = eq[0]
    # There are n = (len(eq)-1) locations, and 2 operation options for each, so n**2 options to deal with... 
    # work in base 2 by using offset of 2^(i+1)
    for i,val in enumerate(eq[1:]):
        offset = 2**(i+1)
        keys = list(q.keys())
        for k in keys:
            k2 = k+offset
            #print(val,k,k2)
            q[k2] = q[k]*val    
            q[k] += val
            #print(q)
    #print(ans, ans in q.values())
    if ans in q.values():
        good.append((ans, eq))
```

```python
final = sum(x for (x,y) in good)
```

```python
%%time
good = []
for ans, eq in data:
    q = {}
    #print(ans, eq)
    q[0] = eq[0]
    # Now giving three options for each step
    # work in base 3 by using offset of 3^(i+1)
    # also adding early termination if > ans
    for i,val in enumerate(eq[1:]):
        offset = 3**(i+1)
        keys = list(q.keys())
        for k in keys:
            k2 = k+offset
            k3 = k + 2*offset
            #print(val,k,k2)
            q[k3] = int(f"{q[k]}{val}")
            if q[k3] > ans:
                del q[k3]                
            q[k2] = q[k]*val
            if q[k2] > ans:
                del q[k2]                
            q[k] += val
            if q[k] > ans:
                del q[k]
            #print(q)
    #print(ans, ans in q.values())
    if ans in q.values():
        good.append((ans, eq))
```

```python
final = sum(x for (x,y) in good)
final
```
