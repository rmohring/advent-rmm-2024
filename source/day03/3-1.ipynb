```python
fh = open('/home/rmohring/code/rmm/advent-rmm-2024/source/day02/3-1.dat')
data = []
sum = 0
enabled = True

for line in fh.readlines():
    tmp = line.split("mul(")
    for xstr in tmp:
        test = re.match(r'[0-9]+,[0-9]+\)', xstr)
        print(f"{xstr} : {test}")
        if enabled and (test is not None):
            newstr = test.group()[:-1]
            a,b = newstr.split(',')
            print(f"{a},{b}")
            mult = int(a)*int(b)
            sum += mult
        if enabled:
            look_for_dont = re.search(r"don\'t\(\)", xstr)
            if look_for_dont is not None:
                enabled = False
        else:
            look_for_do = re.search(r"do\(\)", xstr)
            enabled = (look_for_do is not None)          
fh.close()
```
