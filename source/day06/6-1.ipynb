```python
g = grid.Gridder(data)
```

```python
vec = (0,-1)
```

```python
guard_loc = np.argwhere(g.grid=="^")[0]
guard_loc
```

```python
done=False
rnew=guard_loc[0]
cnew=guard_loc[1]
vectors = [(0,-1),(1,0),(0,1),(-1,0)]
vidx = 0
print(vectors[vidx])
while not done:
    print(rnew, cnew, vidx, vectors[vidx])
    #time.sleep(0.1)
    if (rnew<0) or (rnew>g.grid.shape[0]) or (cnew<0) or (cnew>g.grid.shape[1]):
        done = True
        continue
    if g.grid[rnew, cnew] in ['.','X','^']:
        g.grid[rnew, cnew] = 'X'
        rnew = rnew + vectors[vidx][1]
        cnew = cnew + vectors[vidx][0]
    elif g.grid[rnew, cnew] == "#":
        rnew = rnew - vectors[vidx][1]
        cnew = cnew - vectors[vidx][0]
        vidx += 1
        vidx = vidx%4
    
    
        
```
