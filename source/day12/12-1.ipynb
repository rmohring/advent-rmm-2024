```python
rev = {}
for k,v in g.regions.items():
    if v not in rev:
        rev[v] = [k]
    else:
        rev[v].append(k)                
lens = {}
for k,v in rev.items():
    lens[k] = len(v)
```

```python
lens
```

```python
costs = []
for region in lens.keys():
    area = lens[region]
    perimeter = 0
    test = rev[region][0]
    val = g.val(test)
    print(region, test, val)    
    for p in rev[region]:
        perimeter += 4-len(g.nns[p].get(val, []))
    cost = area * perimeter
    costs.append(cost)
```
